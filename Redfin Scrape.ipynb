```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.redfin.com/CA/San-Francisco/94-Midcrest-Way-94131/home/1975910'
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

response = requests.get(url, headers=headers)

print(response)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('span', class_="entryItemContent")
```

```python
print(results)
```

```python
original_price=None
for result in results:
    try:
        index = result.text.find("Original Price: $")
        if index != -1:
            price=result.text[17:len(result.text)]
            print(price)
            original_price=float(price.replace(',',''))
            print("Original Price: " + (str(original_price)))
                  
            break
      
    except AttributeError as e:
            print(e)
```

```python
listing_date=None
for result in results:
    try:
        search_string = "On Market Date: "
        index = result.text.find(search_string)
        if index != -1:
            listing_date=result.text[len(search_string):len(result.text)]
            print("Listing Date: " + (str(listing_date)))
                  
            break
      
    except AttributeError as e:
            print(e)
```

```python
selling_date=None
for result in results:
    try:
        search_string = "Selling Date:"
        print(result)
        index = result.text.find(search_string)
        if index != -1:
            selling_date=result.text[len(search_string):len(result.text)]
            print("Selling Date: " + (str(selling_date)))
                  
            break
      
    except AttributeError as e:
            print(e)
```
